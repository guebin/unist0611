```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
import pywt

# 방법 1: PyWavelets 라이브러리 사용
def plot_db4_pywt():
    # Daubechies-1 웨이블릿 함수 생성
    wavelet = pywt.Wavelet('db1')
    phi, psi, x = wavelet.wavefun(level=8)
    
    plt.figure(figsize=(12, 5))
    
    # 스케일링 함수 (아버지 웨이블릿)
    plt.subplot(1, 2, 1)
    plt.plot(x, phi, 'b-', linewidth=2)
    plt.title('Harr father wavelet (φ)')
    plt.xlabel('')
    plt.ylabel('')
    plt.grid(True, alpha=0.3)
    
    # 웨이블릿 함수 (어머니 웨이블릿)
    plt.subplot(1, 2, 2)
    plt.plot(x, psi, 'r-', linewidth=2)
    plt.title('Harr mother wavelet (ψ)')
    plt.xlabel('')
    plt.ylabel('')
    plt.grid(True, alpha=0.3)
    
    plt.tight_layout()
#    plt.show()
    plt.savefig("daubechies4.png")

# # 방법 2: 수동으로 계수를 사용하여 구현
# def plot_db4_manual():
#     # Daubechies-4 필터 계수
#     h = np.array([
#         0.6830127, 1.1830127, 0.3169873, -0.1830127,
#         -0.0546875, 0.0312500, 0.0078125, -0.0078125
#     ])
    
#     # 스케일링 함수 계산 (반복적 방법)
#     def scaling_function(h, iterations=10):
#         n = len(h)
#         x = np.linspace(0, n-1, 2**(iterations+1))
#         phi = np.zeros_like(x)
#         phi[len(phi)//2] = 1  # 초기값
        
#         for i in range(iterations):
#             phi_new = np.zeros_like(phi)
#             for k in range(len(h)):
#                 idx = 2*np.arange(len(phi)) - k
#                 valid = (idx >= 0) & (idx < len(phi))
#                 phi_new[valid] += h[k] * phi[idx[valid]]
#             phi = phi_new * np.sqrt(2)
            
#         return x, phi
    
#     # 웨이블릿 함수 계산
#     def wavelet_function(h):
#         g = np.zeros_like(h)
#         for k in range(len(h)):
#             g[k] = (-1)**k * h[len(h)-1-k]
#         return g
    
#     x, phi = scaling_function(h)
#     g = wavelet_function(h)
#     x_psi, psi = scaling_function(g)
    
#     plt.figure(figsize=(12, 5))
    
#     # 스케일링 함수
#     plt.subplot(1, 2, 1)
#     plt.plot(x, phi, 'b-', linewidth=2)
#     plt.title('Daubechies-4 스케일링 함수 (φ)')
#     plt.xlabel('시간')
#     plt.ylabel('진폭')
#     plt.grid(True, alpha=0.3)
    
#     # 웨이블릿 함수
#     plt.subplot(1, 2, 2)
#     plt.plot(x_psi, psi, 'r-', linewidth=2)
#     plt.title('Daubechies-4 웨이블릿 함수 (ψ)')
#     plt.xlabel('시간')
#     plt.ylabel('진폭')
#     plt.grid(True, alpha=0.3)
    
#     plt.tight_layout()
#     plt.show()
```
